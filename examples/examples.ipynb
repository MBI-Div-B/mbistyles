```python
import numpy as np
import matplotlib.pyplot as plt
import mbistyles
```

Define some default plots

```python
def draw_figure(num=8, filename='', grid=False):
    plt.figure()

    x2plot = np.r_[0:15:0.01]
    phases = np.arange(num)*np.pi/4
    amplitudes = 1/(np.arange(num)+1)

    for amplitude, phase in zip(amplitudes, phases):    
        plt.plot(x2plot, amplitude*np.sin(x2plot-phase), label=fr"$\phi = {np.int16(phase/np.pi*4):d}\pi/4$") # , label=f"φ = {phase:d}π/4"

    plt.legend()
    plt.title('MBIstyles')
    plt.xlabel('Delay, $t$ (ps)')
    plt.ylabel('Intensity (arb. units)')

    plt.grid(grid)
    plt.xlim(-0.6, 28.6)
    plt.ylim(-1.15, 1.15)
    
    if len(filename) > 0:
        plt.savefig(f'export/{filename}.svg')
        plt.savefig(f'export/{filename}.png')
        plt.savefig(f'export/{filename}.pdf')
    plt.show()


```

# base style

```python
with plt.style.context('mbistyles.base'):
    draw_figure()
```

# base + paper

```python
with plt.style.context(['mbistyles.base',
                        'mbistyles.paper']):
    draw_figure(filename='base_paper')
```

# 2D

```python
delta = 0.025
x = y = np.arange(-3.0, 3.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2
```

```python
with plt.style.context(['mbistyles.base',
                        'mbistyles.paper']):

    plt.figure()
    plt.imshow(Z, interpolation='bilinear', cmap="RdBu_r",
               origin='lower', extent=[-3, 3, -3, 3],
               vmax=abs(Z).max(), vmin=-abs(Z).max())

    plt.title('MBIstyles')
    plt.xlabel('$x$ (µm)')
    plt.ylabel('$y$ (µm)')
    plt.colorbar()
    plt.show()
```

```python
with plt.style.context(['mbistyles.base',
                        'mbistyles.paper',
                        'mbistyles.ticksinside',
                        'mbistyles.tickseverywhere']):

    plt.figure()
    plt.imshow(Z, interpolation='bilinear', cmap="RdBu_r",
               origin='lower', extent=[-3, 3, -3, 3],
               vmax=abs(Z).max(), vmin=-abs(Z).max())

    plt.title('MBIstyles')
    plt.xlabel('$x$ (µm)')
    plt.ylabel('$y$ (µm)')
    plt.colorbar()
    plt.show()
```

```python
with plt.style.context(['mbistyles.base',
                        'mbistyles.dark',
                        'mbistyles.paper']):

    plt.figure()
    plt.imshow(Z, interpolation='bilinear', cmap="berlin",
               origin='lower', extent=[-3, 3, -3, 3],
               vmax=abs(Z).max(), vmin=-abs(Z).max())

    plt.title('MBIstyles')
    plt.xlabel('$x$ (µm)')
    plt.ylabel('$y$ (µm)')
    plt.colorbar()
    plt.show()
```
