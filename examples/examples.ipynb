```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
from os import path
```

Define some default plots

```python
x2plot = np.r_[0:100:1]
y2plot = np.sqrt(x2plot)
y2plot_scatter = y2plot + (np.random.rand(len(x2plot))-0.5)


def draw_figure(save=False):
    plt.figure()
    plt.plot(x2plot, y2plot_scatter, 'o', label='data')
    plt.plot(x2plot, y2plot_scatter+4, 'o', label='data 2')
    plt.plot(x2plot, y2plot, label='fit', lw=2)
    plt.legend()
    plt.title('Data Plot')
    plt.xlabel('x2plot')
    plt.ylabel('y2plot')
    now = time()
    if save:
        plt.savefig(f'{now}.svg')
        plt.savefig(f'{now}.png')
        plt.savefig(f'{now}.pdf')
    plt.show()


def draw_figure_error(save=False):
    plt.figure()
    plt.errorbar(x2plot, y2plot_scatter, yerr=(np.random.rand(len(x2plot))),
                 fmt='o', alpha=0.5, capsize=2, label='data')
    plt.plot(x2plot, y2plot, label='fit', lw=2)
    plt.legend()
    plt.title('Data Plot Errors')
    plt.xlabel('x2plot')
    plt.ylabel('y2plot')
    if save:
        plt.savefig('test.svg')
        plt.savefig('test.png')
        plt.savefig('test.pdf')
    plt.show()
```

# base style

```python
with plt.style.context(path.abspath('../mbi_base.mplstyle')):
    draw_figure()
```

# base + paper

```python
with plt.style.context(['../mbi_base.mplstyle', './mbi_paper.mplstyle']):
# plt.style.use(['mbi_base.mplstyle'])#, 'mbi_paper.mplstyle'])
    draw_figure(save=True)
    draw_figure_error()
```
