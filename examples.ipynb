```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
```

```python
x2plot = np.r_[0:100:1]
y2plot = np.sqrt(x2plot)
y2plot_scatter = y2plot + (np.random.rand(len(x2plot))-0.5)

def draw_figure():
    plt.figure()
    plt.plot(x2plot, y2plot_scatter, 'o', label='data')
    plt.plot(x2plot, y2plot_scatter+4, 'o', label='data 2')
    plt.plot(x2plot, y2plot, label='fit')
    plt.legend()
    plt.title('Data Plot')
    plt.xlabel('x2plot')
    plt.ylabel('y2plot')
    plt.show()


def draw_figure_error():
    plt.figure()
    plt.errorbar(x2plot, y2plot_scatter, yerr=(np.random.rand(len(x2plot))-0.5),
                 fmt='o', alpha=0.5, capsize=2, label='data')
    plt.plot(x2plot, y2plot, label='fit')
    plt.legend()
    plt.title('Data Plot Errors')
    plt.xlabel('x2plot')
    plt.ylabel('y2plot')
    plt.show()
```

# base style

```python
plt.style.use(['./mbi_base.mplstyle'])

draw_figure()
```

# base + paper

```python
plt.style.use(['./mbi_base.mplstyle', './mbi_paper.mplstyle'])

draw_figure()

draw_figure_error()
```
